This block creates a **histogram** to visualize the distribution of `Avg_Daily_Usage_Hours` among students. It helps us understand how daily social media usage hours are spread across the student population and identify common usage patterns or outliers.

```python
plt.figure(figsize=(10, 6))
plt.hist(df['Avg_Daily_Usage_Hours'], bins=20, edgecolor='black')
plt.title('Distribution of Daily Usage Hours')
plt.xlabel('Hours per day')
plt.ylabel('Frecuency') 
plt.show()
```

This code generates a **bar chart** illustrating the most frequently used social media platforms among students, based on the `Most_Used_Platform` variable. This visualization helps us identify which platforms are dominant within the surveyed student population.

```python
platform_counts = df['Most_Used_Platform'].value_counts()
plt.figure(figsize=(12, 6))
platform_counts.plot(kind='bar')
plt.title('Most Used Platforms')
plt.xlabel('Platform')
plt.ylabel('Number of Students')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

This block creates a **scatter plot** to explore the relationship between students' `Avg_Daily_Usage_Hours` and their `Addicted_Score`. It visually indicates if there's a trend, and then calculates and prints the **Pearson correlation coefficient** to quantify the strength and direction of this relationship, helping us understand the link between usage and addiction.

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['Avg_Daily_Usage_Hours'], df['Addicted_Score'], alpha=0.5)
plt.title('Relationship between Average Daily Usage Hours and Addicted Score')
plt.xlabel('Average Daily Usage Hours')
plt.ylabel('Addicted Score')
plt.show()

# CorrelaciÃ³n
correlation = df['Avg_Daily_Usage_Hours'].corr(df['Addicted_Score'])
print(f'The correlation between the average daily usage hours and the addicted score is: {correlation:.2f}')
```
